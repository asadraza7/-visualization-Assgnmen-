```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x



```python
#Ans:-
x = np.linspace(0, 10, 100) 
y = np.sin(x) 
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)
p.line(x, y, legend_label='sin(x)', line_width=2, line_color='blue')
output_notebook()
show(p)
```

Q(2)Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the 
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
x = np.random.uniform(0, 10, size=50)
y = np.random.uniform(0, 10, size=50)
sizes = np.random.randint(5, 20, size=50)
colors = ['#%06X' % np.random.randint(0, 0xFFFFFF) for _ in range(50)]
p = figure(title="Scatter Plot", x_axis_label='x', y_axis_label='y', width=800, height=400)
p.scatter(x, y, legend_label='Scatter', size=sizes, color=colors, alpha=0.5)
output_notebook()
show(p)
```

Q(3)Generate a Bokeh bar chart representing the counts of different fruits using the following dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] 
counts = [20, 25, 30, 35]


```python
#Ans:-
from bokeh.models import ColumnDataSource
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts', width=800, height=400)
p.vbar(x='fruits', top='counts', source=source, width=0.9, color='navy', alpha=0.7)
output_notebook()
show(p)
```

Q(4)Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000) 
hist, edges = np.histogram(data_hist, bins=30)


```python
#Ans:-
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
source = ColumnDataSource(data=dict(top=hist, left=edges[:-1], right=edges[1:]))
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', width=800, height=400)
p.quad(top='top', bottom=0, left='left', right='right', source=source, fill_color='navy', line_color='white', alpha=0.7)
output_notebook()
show(p)
```
